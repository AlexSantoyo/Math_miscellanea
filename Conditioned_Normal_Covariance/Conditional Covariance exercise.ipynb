**Problem**

Consider $X$ and $Y$ iid standard Normal random variables, so $X\stackrel{d}{=}Y\sim N(0,1)$.
Calculate the covariance of $X$ and $Y$ given that $X+Y>0$.

Some facts we may use later are as follows.

The density of $X$ is given by
$$ f_X(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}.$$

So, notice the following property:
$$ f^{\prime}_X(x) = \frac{1}{\sqrt{2\pi}}x e^{-\frac{x^2}{2}} = xf_X(x).$$
Which can be used to prove the following property of a Normal rv.

The expected value of $X \sim N(\mu,\sigma^2)$ conditioned to be on a given inteval $(a,b)$ is given by
$$ \mathbb{E}(\left. X \right| a<X<b) = \mu - \sigma^2 \frac{f(b)-f(a)}{F(b)-F(a)},$$
where $f$ and $F$ are the density and the distribution function of $X$, respectively.

In particular, if $X \sim N(0,2)$, then:
$$ \mathbb{E}(\left. X \right| a<X) = - 2 \frac{f(\infty)-f(a)}{F(\infty)-F(a)} = \frac{2f(a)}{1-F(a)}.$$


Let's consider the following rvs:
$$
\begin{align*}
    W &= \left. X \right| X+Y>0,\\
    Z &= \left. Y \right| X+Y>0.
\end{align*}
$$

We aim to calculate:
$$
Cov(W,Z) = \mathbb{E}(WZ) - \mathbb{E}(W)\mathbb{E}(Z).
$$

$\require{cancel}$
$\textbf{Direct calculations}$

Using the definition of conditional expectation, total probability theorem and the density of $X$ and $Y$ we have:
$$
\begin{align*}
    \mathbb{E}(W) &= \mathbb{E}(\left. X \right| X+Y>0) = \frac{\mathbb{E}( X  1_{\{X+Y>0\}})}{\mathbb{P}(X+Y>0)}\\
    &= 2\int_{-\infty}^{\infty} \mathbb{E}( X 1_{\{X+Y>0\}}) f_Y(y)dy = 2\int_{-\infty}^{\infty} \mathbb{E}( X 1_{\{X>-y\}}) f_Y(y)dy \\ 
    &= 2\int_{-\infty}^{\infty} \left( \int_{-y}^{\infty} x f_X(x)dx \right) f_Y(y) dy \\
    &= \frac{1}{\pi}\int_{-\infty}^{\infty} \int_{-y}^{\infty} x e^{-\frac{1}{2}(x^2 + y^2)} dy
\end{align*}
$$
Now we can use polar coordinates to make the calculation of the integral easier. Let $x = r\cos(\theta)$, $y = r\sin(\theta)$ and recall that the Jacobian of the transformation is given by $r$, since
$$ 
J = \begin{vmatrix}
\cos(\theta) & \sin(\theta)\\
-r\sin(\theta) & r\cos(\theta) 
\end{vmatrix}
= r.
$$
Since we are in the case $x+y>0$, we can see that this is equivalent to consider $r>0$ and $\theta \in [-\pi/4,3\pi/4]$.

Then, continuing with the calculations we make the change of coordinates, then using integration by parts, properties of the normal density and trigonometric identities we have:
$$
\begin{align*}
    \mathbb{E}(W) &= \frac{1}{\pi}\int_{-\infty}^{\infty} \int_{-y}^{\infty} x e^{-\frac{1}{2}(x^2 + y^2)} dy\\
    &= \frac{1}{\pi}\int_{-\frac{\pi}{4}}^{\frac{3\pi}{4}} \int_{0}^{\infty} r\cdot r\cos(\theta) e^{-\frac{r^2}{2}} dr d\theta\\
    &= \frac{1}{\pi}\int_{-\frac{\pi}{4}}^{\frac{3\pi}{4}} \cos(\theta) \left( \int_{0}^{\infty} r\cdot r e^{-\frac{r^2}{2}} dr \right) d\theta\\
    &= \frac{1}{\pi}\int_{-\frac{\pi}{4}}^{\frac{3\pi}{4}} \cos(\theta) \left(\cancelto{0}{ \left. - r e^{-\frac{r^2}{2}} \right|_{0}^{\infty}} + \sqrt{2\pi} \cancelto{1/2}{\frac{1}{\sqrt{2\pi}}\int_{0}^{\infty} e^{-\frac{r^2}{2}} dr} \right) d\theta\\
    &= \frac{\sqrt{2\pi}}{2\pi}\int_{-\frac{\pi}{4}}^{\frac{3\pi}{4}} \cos(\theta) d\theta = \frac{\sqrt{2\pi}}{2\pi} \left. \sin(\theta)\right|_{-\pi/4}^{3\pi/4}\\
    &= \frac{\sqrt{2\pi}}{2\pi} \left[ \sin(3\pi/4) - \sin(-\pi/4) \right] = \frac{\sqrt{2\pi}}{2\pi} \frac{2}{\sqrt{2}}\\
    &= \frac{1}{\sqrt{\pi}}.
\end{align*}
$$

In summary, we have that $\mathbb{E}(W) = \mathbb{E}(Z) = \frac{1}{\sqrt{\pi}}$.

In a similar fashion, we can calculate $\mathbb{E}(WZ)$. By geometric inspection it is easy to see that this expectation corresponds to the integral of a function which is symmetric around the identity $y=f(x)=x$ and thus it is zero, but let's compute the integral as a sanity check.
$$
\begin{align*}
    \mathbb{E}(WZ) &= \frac{\mathbb{E}( XY  1_{\{X+Y>0\}})}{\mathbb{P}(X+Y>0)}\\
    &= 2\int_{-\infty}^{\infty} \mathbb{E}( Xy 1_{\{X>-y\}}) f_Y(y)dy \\ 
    &= 2\int_{-\infty}^{\infty}\int_{-y}^{\infty} xy f_X(x)f_Y(y) dxdy \\
    &= \frac{1}{\pi}\int_{-\infty}^{\infty} \int_{-y}^{\infty} xy e^{-\frac{1}{2}(x^2 + y^2)} dxdy
\end{align*}
$$
Again, using polar coordinates and the same integration by parts and trigonometric identities, we have:
$$
\begin{align*}
    \mathbb{E}(WZ) &= \frac{1}{\pi}\int_{-\infty}^{\infty} \int_{-y}^{\infty} xy e^{-\frac{1}{2}(x^2 + y^2)} dxdy\\
    &= \frac{1}{\pi}\int_{-\pi/4}^{3\pi/4} \int_{0}^{\infty} r^2\sin(\theta)\cos(\theta) e^{-\frac{r^2}{2}} dr d\theta\\
    &= \frac{1}{\pi}\int_{-\frac{\pi}{4}}^{\frac{3\pi}{4}} \sin(\theta)\cos(\theta) \left( \int_{0}^{\infty} r\cdot r e^{-\frac{r^2}{2}} dr \right) d\theta\\
    &= \frac{\sqrt{2\pi}}{2\pi}\int_{-\frac{\pi}{4}}^{\frac{3\pi}{4}} \sin(\theta)\cos(\theta) d\theta\\
    &= \frac{\sqrt{2\pi}}{2\pi} \left. \frac{\sin^2(\theta)}{2}\right|_{-\pi/4}^{3\pi/4}\\
    &= \frac{\sqrt{2\pi}}{2\pi} \left[ \frac{\sin^2(3\pi/4)}{2} - \frac{\sin^2(-\pi/4)}{2} \right] = \frac{\sqrt{2\pi}}{4\pi} \left[ \frac{1}{2} - \frac{1}{2} \right]\\
    &= 0.
\end{align*}
$$

Finally, we get:
$$
Cov(W,Z) = \mathbb{E}(WZ) - \mathbb{E}(W)\mathbb{E}(Z) = 0 - \left(\frac{1}{\sqrt{\pi}}\right)^2 = -\frac{1}{\pi}
$$

**Shorter calculation**

Since $X$ and $Y$ are iid standard Normal r.v., we have that $X+Y \sim N(0,2)$. And since expectation is linear we have:
$$
\begin{align*}
\mathbb{E}(\left. X \right| X+Y>0) + \mathbb{E}(\left. Y \right| X+Y>0) &=  \mathbb{E}(\left. X + Y \right| X+Y>0)\\
&= \frac{2f_{X+Y}(0)}{1-F_{X+Y}(0)} = \frac{2/\sqrt{4\pi}}{1/2} \\
&= \frac{4}{\sqrt{4\pi}}= \frac{2}{\sqrt{\pi}}.
\end{align*}
$$
This implies that:
$$
\mathbb{E}(\left. X \right| X+Y>0) = \mathbb{E}(\left. Y \right| X+Y>0) = \frac{\mathbb{E}(\left. X + Y \right| X+Y>0)}{2} = \frac{1}{\sqrt{\pi}}.
$$
Finally, arguing by symmetry $\mathbb{E}(\left. XY \right| X+Y>0)=0$, so the covariance is calculated as:
$$
Cov(W,Z) = \mathbb{E}(WZ) - \mathbb{E}(W)\mathbb{E}(Z) = 0 - \left(\frac{1}{\sqrt{\pi}}\right)^2 = -\frac{1}{\pi}.
$$

**Simulating rvs in python**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(666)
# generate samples for X and Y
X= np.random.normal(0,1,size=1000000)
Y= np.random.normal(0,1,size=1000000)
# save them in a dictionary to easy manipulation and plots
xy_dict = {'X':X, 'Y':Y}
xy_df = pd.DataFrame(xy_dict)
print(xy_df.head(5))
```

```python
plt.scatter(xy_df.X,xy_df.Y,alpha=0.5)
plt.plot([min(xy_df.X),max(xy_df.Y)],[max(xy_df.X),min(xy_df.Y)], c='red',label='x+y=0')
plt.legend()
plt.gca().set_aspect(1.0)
plt.show()
```

```python
# adding the values to filter later
xy_df['XplusY']=xy_df.X + xy_df.Y
xy_df['XY']=xy_df.X * xy_df.Y

# values for W and Z, that is X and Y filtered by X+Y=0
wz_df = xy_df[xy_df.XplusY>0]
print(f"The ratio of the size of the filtered values {wz_df.shape[0]/xy_df.shape[0]}, should be near 0.5")
# renaming columns
wz_df.columns = ["W","Z","WplusZ","WZ"]
print(wz_df.head())
```

```python
W_mean = np.mean(wz_df.W)
Z_mean = np.mean(wz_df.Z)
WZ_mean = np.mean(wz_df.WZ)
cov_cond = WZ_mean - W_mean*Z_mean
print(f"The covariance is equal to {cov_cond}")
print(f"And should be close -(1/pi) = {-1/(np.pi)}")
```
